```python
gene = []
#extrayendo los nombres de los genes abreviados de el archivo excel
for i in range(11):
    if i == 0:
        data = pd.read_excel("Genes identificados con vecinos.xlsx",
                    sheet_name=i,
                     index_col="N°",
                    skiprows=4)
    else:
        data = pd.read_excel("Genes identificados con vecinos.xlsx",
                        sheet_name=i,
                         index_col="N°",
                        skiprows=1)
    data.to_csv("genes.csv")
    
    ef = pd.read_csv('genes.csv')
    gene.extend(list(ef["Gen Abrev"]))
gene1 =[]
#print(len(gene))
for i in range(35):
    gene1.append(genes[i])
    
Entry = []
Entry_name = []
Protrein = []
Gene = []
Organism = []
Status = []
GO_Biologicalprocess = []

#gene1 = tiene que ser la lista con todos los genesabrev
for i in gene1:
    URL=str(i) 
    #formando el enlace  la pagina que vamos escrapear 
    URL= "https://www.uniprot.org/uniprot/?query="+i+"+Bos+taurus&sort=score"
    html = requests.get(URL)
    soup = BeautifulSoup(html.content, 'html.parser')
    
    
    
    c =soup.find("table",{"class":"grid"})
    
    if c != None : 
        entry = c.findAll("td" , {"class":"entryID"})[0].find("a").text
        entry_name = c.findAll("td")[3].text
        ##########
        #Entry 
        ##########
        Entry.append(entry)
        ############
        #Entry_name 
        ############
        Entry_name.append(entry_name)
        
        url = "https://www.uniprot.org/uniprot/"+entry
        html = requests.get(url)
        soup = BeautifulSoup(html.content, 'html.parser')
       
        f =soup.find("div",{"id":"entry-overview"})
        #################
        #Protrein
        #################
        protrein = f.find("div" , {"id":"content-protein"}).find("h1").text
        Protrein.append(protrein)
        #################
        #Gene
        #################
        gene = f.find("div", {"id":"content-gene"}).find("h2").text
        Gene.append(gene)
        #################
        #Organism
        #################
        organism = f.find("div", {"id":"content-organism"}).find("em").text
        Organism.append(organism)
        ################
        #Status
        ###############
        status = f.find("div", {"id":"content-status"}).find("a").find("span").text
        Status.append(status)
        #####################
        #Go_Biologicalprocess
        #####################
        page=soup.find("ul",{"class":"noNumbering biological_process"}) 
        
        if page==None:
            GO_Biologicalprocess.append("Na")
        else:
            location=page.find_all("a",{"onclick":"window.ga('UniProt-Entry-View', 'click', 'Display-GO-Term');"})
            cadena = ""
            for i in range(0,len(location)):
                if location[i].text=="":
                    cadena = cadena + "-"
                else:
                    cadena = cadena + location[i].text + "."
                
            GO_Biologicalprocess.append(cadena)               
    else :
        GO_Biologicalprocess.append("Na")
        Entry.append("Na")
        Entry_name .append("Na")
        Protrein.append("Na")
        Gene.append("Na")
        Organism.append("Na")
        Status.append("Na")


print(Entry)   
print(Entry_name)
print(Protrein)
print(Gene)
print(Organism)
print(Status)
print(GO_Biologicalprocess)
print(len(GO_Biologicalprocess))
```
